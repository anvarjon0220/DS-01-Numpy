```python
import numpy as np
```

```python
# Massivlar
xarr = np.array([1.1, 1.2, 1.3, 1.4, 1.5]) # xarr
yarr = np.array([2.1, 2.2, 2.3, 2.4, 2.5]) # yarr
```

```python
# Shart
cond = np.array([True, False, True, True, False])
```

```python
# where
result = np.where(cond, xarr, yarr) # cond ==True : xarr, cond==False : yrr
result
```
