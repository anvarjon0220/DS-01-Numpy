```python
import numpy as np
```

```python
arr = np.random.rand(4)
arr
```

# sort funksiyasi

```python
sorted_arr = np.sort(arr) # tartiblangan yangi array
sorted_arr
```

```python
# 2-usul
arr.sort() # asl arrayning tartiblangani
arr
```

```python
# teskari tartiblash
rev_arr = -np.sort(-arr)
rev_arr
```

```python
# 2D array
arr2d = np.random.randn(2, 4)
arr2d
```

```python
# axis = 0 --> Qator
row_srt = np.sort(arr2d, axis=0)
row_srt
```

```python
col_srt = np.sort(arr2d, axis=1)
col_srt
```
